## Handling Categorical Variables

```python
df=pd.get_dummies(df)
```

```python
X=df.drop('quality', axis=1)
Y=df['quality']
```

```python
X
```

```python
Y
```

## Handling Missing Data 

```python
from sklearn.impute import SimpleImputer

X_copy=X.copy()

imputer=SimpleImputer(strategy='median')
imputer.fit(X_copy)
X=imputer.transform(X_copy)
#X
```

```python
X=pd.DataFrame(X, columns=X_copy.columns, index=X_copy.index)
```

```python
X.isnull().sum()
```

```python
X.describe()
```

## Handling Imbalance Dataset

```python
#from imblearn.combine import SMOTETomek

#smk=SMOTETomek()
#x_train, y_train=smk.fit_sample(X,Y)
```

```python
from imblearn.over_sampling import RandomOverSampler

rs= RandomOverSampler()
X, Y=rs.fit_sample(X,Y)
```

```python
print('X=', X.shape)
print('Y=', Y.shape)
print(Y.value_counts())
```

## Splitting data into Train and Test data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test= train_test_split(X, Y, train_size=0.8, random_state=42, stratify=Y)
```

```python
print('x train:',x_train.shape)
print('x test',x_test.shape)
print('y train',y_train.shape)
print('y test', y_test.shape)
```

## Training

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.dummy import DummyClassifier
from sklearn.svm import SVC
```

```python
pipe=Pipeline([('standardscaler', StandardScaler()),('classifier', DummyClassifier())])

grid_param=[
                #{
                #    'classifier': [LogisticRegression(multi_class='multinomial', solver='lbfgs')],
                #    'classifier__penalty':['l2','l1'],
                #    'classifier__C':np.logspace(0,4,10)
                #},
                {
                    'classifier':[SVC()],
                    'classifier__kernel':['rbf'],
                    'classifier__gamma':[0.001,0.0001,0.01,0.1, 0,1,10,100,1000,10000],
                    'classifier__C':[0.001,0.0001,0.01,0.1, 0,1,10,100,1000,10000],
                    #'classifier__degree':np.arange(0,11),
                    #'classifier__coef0':[0,1,2,3,4,5]
                }
                #{
                #    'classifier':[RandomForestClassifier()],
                #    'classifier__n_estimators':np.arange(1,1001),
                #    'classifier__max_leaf_nodes':np.arange(1,11)
                #}
]
```

```python
from sklearn.model_selection import StratifiedKFold

skf=StratifiedKFold(n_splits=5, random_state=None)   
```

```python
grid_search= GridSearchCV(pipe, grid_param, cv=skf, verbose=4, n_jobs=-1, scoring='f1_macro')
best_model=grid_search.fit(x_train, y_train)
```
